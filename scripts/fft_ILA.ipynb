```python
file = '128_point.coe'
with open(file,'r') as f:
    data = f.readlines()
data = data[1].split('=')[1].split(";")[0].split()[:-2]
print(len(data))
data
```

```python
inp_r = list()
inp_i = list()
for i in range(0,len(data),2):
    inp_r.append(int(data[i]))
    inp_i.append(int(data[i+1]))
len(inp_i),len(inp_r)
inp_r = np.array(inp_r)
inp_i = np.array(inp_i)
```

```python
inputs = inp_r/64.0+1j*inp_i/64.0
inputs
```

```python
import pandas as pd
df = pd.read_csv('ila_final_128.csv')
output_r = df['out_case_r[18:0]'][1:]
output_i = df['out_case_i[18:0]'][1:]
df.head()
```

```python
output_i
```

```python
output_r = np.array(output_r[222:350])
output_r = [float(i)/64.0 for i in output_r]
output_i = np.array(output_i[222:350])
output_i = [float(i)/64.0 for i in output_i]
```

```python
outputs = np.array(output_r)+1j*np.array(output_i)
```

```python
import numpy as np
import matplotlib.pyplot as plt
inputs = np.array(inputs)
outputs = np.array(outputs)

# Assuming 'inputs' and 'outputs' are extracted as complex arrays
# Your provided values should already be stored in `outputs`, and Python FFT will be stored in `fft_outputs`

# Perform FFT on the extracted inputs
fft_outputs = np.fft.fft(inputs)

# Plotting the results
plt.figure(figsize=(12, 6))

# Subplot 1: Provided FFT Outputs
plt.subplot(1, 2, 1)
plt.title("Imag FFT Outputs")
plt.plot(fft_outputs.imag, label='Imaginary Part from Python', linestyle='-')
plt.plot(outputs.imag, label='Imaginary Part from Vivado',linestyle='--')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)

# Subplot 2: Computed FFT Outputs (Python)
plt.subplot(1, 2, 2)
plt.title("Real FFT Outputs")
plt.plot(fft_outputs.real, label='Real Part from Python', linestyle='-')
plt.plot(outputs.real, label='Real Part from Vivado', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.title("Relative Error in FFT Outputs")
plt.plot([abs((i-j)/j) for i,j in zip(outputs.real,fft_outputs.real)], label='Relative Error in real parts', linestyle='-')
plt.plot([abs((i-j)/j) for i,j in zip(outputs.imag,fft_outputs.imag)], label='Relative Error in imaginary parts', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()
```
